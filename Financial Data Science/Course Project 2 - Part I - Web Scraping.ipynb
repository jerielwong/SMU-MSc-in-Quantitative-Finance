```python
import urllib
from urllib.request import urlopen
import pandas as pd
from bs4 import BeautifulSoup
import numpy as np
import matplotlib.pyplot as plt
import string as st
import seaborn as sns
import time
```

# Steps

## Step 1 : Scrap for all cities to retrieve full list of employers
## Step 2 : Scrap with full list of employers 

# Step 1 Looping across cities, to get all employers

```python
#preload parameters for scrapping cities names

url = 'https://h1bdata.info/topcities.php'
page = urlopen(url)
data2 = BeautifulSoup(page,'html.parser')

# cities, detecting cities through all a href tag
cities = []
for x in data2.find_all('a', href=True):
    cities.append(x["href"].split("=")[-1])

# hard coded to select cities from row 33, to select relevant cities
cities = cities[33:]

# drop cities that has empty table , or dead link
del cities[1621]
cities.remove("KINGSVILLE")

# =================== Looping each city after getting all cities names ============================

# final empty list to store each observations while cycling across cities
final = []
#loop across cities 
for each_city in cities:
# retrieving labels from the website's table
    r = urllib.request.urlopen('https://h1bdata.info/index.php?em=&job=&city='+each_city+'&year=All+Years')
    soup = BeautifulSoup(r)
    data2 = soup.find_all('tr')    
    labels = []
    for h in data2[0].find_all('th'):
        labels.append(h.get_text().strip().lower())

# after getting labels, i.e. the column titles, this is to extract each row from the table
# and double append so it becaomes a 1x6 dimension

    for data_rows in data2[1:]:
        data_list = []
        for d in data_rows.find_all('td'):
            d_str = d.get_text().replace(',','')
            data_list.append(d_str)
        final.append(data_list)
    print(f'{each_city} is completed')
    time.sleep(1.1)
    
#putting cities results as a df 
df = pd.DataFrame(final, columns = labels)    
df['submit date'] = pd.to_datetime(df['submit date'])
df['start date'] = pd.to_datetime(df['start date'])
df['state'] = df['location'].str.split().str[-1] 
df['year'] = df['submit date'].dt.year
df['month'] = df['submit date'].dt.month
```

# Step 2 Looping across all employers (130k+) via unique values of employer names, from cities scrapped above

```python
# make list for unique values of employers
employers = list(df['employer'].unique())

# add + sign in between citiy name with spaces, so that URL will run
employers = [x.replace(" ","+") for x in employers ]

# remove values in employers list that contain \n
for employer_ind in employers:
    if employer_ind.find("\n")>-1: # locate true by using >-1
        employers.pop(employers.index(employer_ind))

#search for anymore special characters
special_characters = !@#$%^&*()-+?_=,<>/

if any(x1 in special_characters for x1 in employers):
    print(x1.index)
else:
    print("no")
    
#Gather all the employers tables and row
final = []
for each_employer in employers:
# retrieving labels from the website's table
    labels = []
    each_employer = each_employer.encode('ascii','ignore').decode('ascii')  # to handle han yu ping yin 
    r = urllib.request.urlopen('https://h1bdata.info/index.php?em='+each_employer+'&job=&city=&year=All+Years')
    soup = BeautifulSoup(r)
    data2 = soup.find_all('tr')   #scan table rows     
    
    if len(data2) == 0: #if data2 has no values, go to next loop
        continue
    for h in data2[0].find_all('th'): #scan for table headers
        labels.append(h.get_text().strip().lower())

# after getting labels, i.e. the column titles, this is to extract each row from the table
# and double append so it becaomes a 1x6 dimension

    for data_rows in data2[1:]:
        data_list = []
        for d in data_rows.find_all('td'):
            d_str = d.get_text().replace(',','')
            data_list.append(d_str)
        final.append(data_list)
    print(f'{each_employer} is completed')
    time.sleep(1.1)
    
```

```python
#putting it into df
labels = ['employer',
 'job title',
 'base salary',
 'location',
 'submit date',
 'start date']

df = pd.DataFrame(final, columns = labels)    
df['submit date'] = pd.to_datetime(df['submit date'])
df['start date'] = pd.to_datetime(df['start date'])
df['state'] = df['location'].str.split().str[-1] 
df['year'] = df['submit date'].dt.year
df['month'] = df['submit date'].dt.month

df
```
